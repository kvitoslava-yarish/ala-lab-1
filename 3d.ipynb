```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import cv2 as cv
```

```python
class threeDObject:
    def __init__(self, coordinates: np.array):
        self.coordinates = coordinates
        
    def reflection(self, axis:str):
        matrix = np.eye(3)
        if axis == 'x':
            matrix[0][0] = -1
        elif axis == 'y':
            matrix[1][1] = -1
        elif axis == 'z':
            matrix[2][2] = -1
        elif axis == '0':
            matrix = matrix * -1
        else:
            raise ValueError("incorrect axis")
        new_coordinates = self.coordinates @ matrix
        self.print_object(new_coordinates, "reflection")
        return new_coordinates 
    
    def scale(self, factor: float):
        new_coordinates = self.coordinates * factor
        self.print_object(new_coordinates, "scaling")
        return new_coordinates
    
    def print_object(self, coordinates: np.array, title: str):
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        x = coordinates[:, 0]
        y = coordinates[:, 1]
        z = coordinates[:, 2]
        ax.plot(x, y, z, marker='o')
        ax.set_title(title)
        plt.show()
    def __str__(self):
        self.print_object(self.coordinates, 'origin matrix')
        return ""
        
```

```python
triangle = threeDObject(np.array([[1,1,1], [1,2,2], [2,2,2],[1,1,1]]))
```

```python
# x reflection
print(triangle.reflection("x"))
```

```python
# y reflection
print(triangle.reflection("y"))
```

```python
# z reflection
print(triangle.reflection("z"))
```

```python
# 0 reflection
print(triangle.reflection("0"))
```

```python
print(triangle.scale(2))
```
